## Этап 5. Реализуйте подсчет метрик TPR, FPR «вручную», без использования готовых функций из библиотеки sklearn.

```python
import numpy as np 
y_test_1 = []
for i in Y_test:
    for n in i:
        y_test_1.append(n)

y_test = np.asarray(y_test_1)
y_test #истинные значения
```

```python
def ROC_hand(y_pred, y_test):
    tp = 0 # True positive
    fp = 0 # False positive
    fn = 0 # False negative
    tn = 0 # True negative
    for predicted_prob, actual in zip(y_pred[:,1], y_test):
        if predicted_prob >= 0.5:
            predicted = 1
        else:
            predicted = 0

        if predicted == 1:
            if actual == 1:
                tp += 1
            else:
                fp += 1

        else:
            if actual == 1:
                fn += 1
            else:
                tn += 1
    return tp, fp, fn, tn
```

```python
tp_, fp_, fn_, tn_ = ROC_hand(predictions, y_test)
```

```python
#false positive rate
try:
    fpr_ = fp_/(fp_+tn_)
    print(f'fpr: {fpr_}')
except:
    print('Деление на ноль в fpr')

#true positive rate
#Recall - доля правильно предсказанных среди категории 1 (как много медалистов мы нашли?)
try:
    tpr_ = tp_/(tp_+fn_)
    print(f'tpr: {tpr_}')
except:
    print('деление на ноль в tpr')
```

## Этап 6. Постройте ROC-кривую с помощью вычисленных в п. 5 метрик: объедините графики из п. 3 и п. 6 в один. Сравните, сделайте вывод.

```python
arange_ = np.arange(0.00,1.00,0.01)
arange_
```

```python
def ROC_hand_curve(y_pred, y_test, threshold):
    fpr_list = []
    tpr_list = []
    tp = 0 # True positive
    fp = 0 # False positive
    fn = 0 # False negative
    tn = 0 # True negative
    for i in threshold:
        for predicted_prob, actual in zip(y_pred[:,1], y_test):
            if predicted_prob >= i:
                predicted = 1
            else:
                predicted = 0

            if predicted == 1:
                if actual == 1:
                    tp += 1
                else:
                    fp += 1

            else:
                if actual == 1:
                    fn += 1
                else:
                    tn += 1
                    
            try:
                fpr_list.append(fp/(fp+tn))
            except ZeroDivisionError:
                pass
            
            try:
                tpr_list.append(tp/(tp+fn))
            except ZeroDivisionError:
                fpr_list.pop(-1)
                pass
            
    
           
    return tp, fp, fn, tn, fpr_list, tpr_list
```

```python
def presicion_recall_test(y_pred_ver, y_test, threshold):
    tp_fp_fn_tn = []
    tp = 0 # True positive
    fp = 0 # False positive
    fn = 0 # False negative
    tn = 0 # True negative
    for i in threshold:
        for predicted_prob, actual in zip(y_pred_ver[:,1], y_test):
            if predicted_prob >= i:
                predicted = 1
            else:
                predicted = 0

            if predicted == 1:
                if actual == 1:
                    tp += 1
                else:
                    fp += 1

            else:
                if actual == 1:
                    fn += 1
                else:
                    tn += 1
        tp_fp_fn_tn.append([tp, fp, fn, tn])
    return tp_fp_fn_tn
```
